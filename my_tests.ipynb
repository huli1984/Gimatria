```python
import numpy as np
def dic_conv(argument):
    switcher = {"א": 1,
                        "ב": 2,
                        "ג": 3,
                        "ד": 4,
                        "ה": 5,
                        "ו": 6,
                        "ז": 7,
                        "ח": 8,
                        "ט": 9,
                        "י": 10,
                        "כ": 20,
                        "ך": 20,
                        "ל": 30,
                        "מ": 40,
                        "ם": 40,
                        "נ": 50,
                        "ן": 50,
                        "ס": 60,
                        "ע": 70,
                        "פ": 80,
                        "ף": 80,
                        "צ": 90,
                        "ץ": 90,
                        "ק": 100,
                        "ר": 200,
                        "ש": 300,
                        "ת": 400}
    return switcher.get(argument, lambda: "invalid character: use only hebrew characters.")


def convert_list(word_list):
    gim_word = []
    gim_word_list = []
    for word in word_list:
        if word.isdigit():
            gim_word.append(int(word))
            gim_word.append(0)
        else:
            for i in word:
                j = dic_conv(i)
                gim_word.append(int(j))
            summa_parziale = sum(gim_word)
            # print(summa_parziale, "somma parziale")
            gim_word_list.append(summa_parziale)
            gim_word = []
    summa = gim_word_list
    # print(summa, "summa")
    return summa

super_word_index = []
super_booklist = []
db = pd.DataFrame()
index_list_sup = []
splitted_list_sup = []
values_converted_sup = []    
for selected_book in selected_books:
    my_Bible_book = open("Bereshit_lines.txt", "r")
    my_bb_lines = my_Bible_book.readlines()
    
    for my_bb_line in my_bb_lines:
        splitted_line = my_bb_line.split()
        single_index = str(splitted_line[0]).replace("[", "").replace("]", "").replace("\u200f", "")
        
        index_list = [str(splitted_line[0]).replace("[", "").replace("]", "").replace("\u200f", "")]*len(splitted_line[1:])
        splitted_list = splitted_line[1:]
        values_converted = convert_list(splitted_list)
        index_list_sup.extend(index_list)
        splitted_list_sup.extend(splitted_list)
        values_converted_sup.extend(values_converted)
        
        
    db["values"] = values_converted_sup
    db["words"] = splitted_list_sup
    db["line address"] = index_list_sup
    db.set_index("line address")
    
print(db.head(), db.count(), db[db["values"] == 246])
```
